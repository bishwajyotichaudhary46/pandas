```python
d1={
'key1':[1,2,3,4,5,6],
    'key2':[7,8,9,10,11,12],
    'key3': [13,14,15,16,17,18],
    'key4': [19,20,21,22,23,24]

}
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(d1,index = ['a','b','c','d','e','f'])
```

```python
df1
```

```python
d2={
'key1':[1,2,3,4,5,"sudh"],
    'key2':[7,8,"kumar",10,11,12],
    'key3': [13,14,15,16,17,18],
    'key4': [19,20,21,"ineuron","pwskills",24]

}
```

```python
df2 = pd.DataFrame(d2)
```

```python
df2
```

```python
# default inner join --> intersection
pd.merge(df1, df2)
```

```python
# merge insence of inner , left,join
# on check only one column like key1 intersection and other same
pd.merge(df1,df2, how = "inner", on="key1")
```

```python
# only row which match in both dataframe in key4
pd.merge(df1,df2, how = "inner", on="key4")
```

```python
# try to take  every data  from left side
pd.merge(df1,df2, how = "left", on="key4")
```

```python
# try to take  every data  from left side but it does not take right side but incase of on --> it give intertion of column
pd.merge(df1,df2, how = "left")
```

```python
# it take every thing from right table , on--> which take intersection on both put match of data from left
pd.merge(df1,df2,how = "right", on="key4")
```

```python
# it every thing both table 
pd.merge(df1,df2,how = "outer", on="key4")
```

```python
# it every thing both table, all things  union
pd.merge(df1,df2,how = "outer")
```
